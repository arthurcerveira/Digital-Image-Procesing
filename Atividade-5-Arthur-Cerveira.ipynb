## Exercício 1: Transformada de Fourier

O exercício 1 consiste em aplicar a Transformada de Fourier em uma imagem e exibir o resultado. A imagem original está apresentada abaixo.

```octave
lenna = imread('./images/lena_cinza.bmp');
imshow(lenna);
```

```octave
transform = fft2(lenna);

imshow(log(abs(transform)), []);
```

```octave
result = log(abs(fftshift(transform)));

imshow(result);
```

## Exercício 2: Função de transformadas

O exercício 3 consiste em implementar uma função que executa as funções de Trandoformada e Transformada Inversa de acordo com o valor passado no argumento `s`. Observou-se que, embora as imagens sejam exibidas corretamente, o resultado das imagens salvas não está adequado, sendo apenas ruído para a imagem transformada, e uma imagem em branco para a imagem original obtida a partir da transformada inversa. 

```octave
function img_transform = transform_freq(imagem, s)
    if (s == 0)
        transform = fft2(imagem);
        
        img_transform = log(abs(fftshift(transform)));

        imwrite(transform, "./images/transform_lenna.jpg");

        imshow(img_transform, []);
    endif
    
    if (s == 1)
        original = ifft2(ifftshift(imagem));
        
        imwrite(original, "./images/original_lenna.jpg");

        imshow(original, []);
    endif
endfunction
```

```octave
transform_freq(original, 0);
```

```octave
result = imread('./images/transform_lenna.jpg');
imshow(result);
```

```octave
transform_freq(fftshift(transform), 1);
```

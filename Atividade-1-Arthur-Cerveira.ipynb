## Exercício 1 - Manipulando uma imagem

O exercício 1 consiste em uma manipulação simples de imagens, onde uma imagem deve ser lida, convertida para outro espaço de cores, e modificada através de operações simples. Foi escolhida imagem da Lenna para a realização do exercício.

```octave
lenna = imread('./images/Lenna.png');
imshow(lenna);
```

No exercício é pedido que a imagem seja convertida do espaço de cores RGB para YCbCr, e então sejam exibidos cada um dos canais da imagem. Aqui é possível ver como o canal de luminância guarda mais informações do que os canais de crominância. Para realizar essa conversão foi necessário carregar o pacote `image`.

```octave
pkg load image

YCbCr = rgb2ycbcr(lenna);

imshow(YCbCr(:, :, 1));
```

```octave
imshow(YCbCr(:, :, 2));
```

```octave
imshow(YCbCr(:, :, 3));
```

Em seguida foi traçado um histograma da luminância da imagem. Então a imagem é escurecida e é traçado o histograma novamente, onde podemos visualizar como valor máximo dimunui de 250 para 200. Esse canal é então salvo como um arquivo PNG.

```octave
hist(YCbCr(:, :, 1));
```

```octave
YCbCr(:, :, 1) = YCbCr(:, :, 1) * 0.8;

hist(YCbCr(:, :, 1));

imwrite(YCbCr(:, :, 1), './images/dark_lenna.png');
```

A última da tarefa do exercício 1 consiste em realizar uma subamostragem dos canais de crominância da imagem. Essa operação foi realizada através da indexação com passo 2, indo desde a primeira posição até as extremidades da imagem. É possível observar como subamostragem tem apenas metade da largura e altura da imagem original, além de ter apenas um canal.

```octave
[W, H, D] = size(YCbCr);

Cb = YCbCr(1:2:W, 1:2:H, 2);
Cr = YCbCr(1:2:W, 1:2:H, 3);

size(YCbCr)
size(Cb)
```

## Exercício 2 - Modificando cores

Esse exercício consiste em modificar as cores da imagem de uma borboleta e salvá-la com outras dimensões escolhidas pelo usuário.

```octave
butterfly = imread('./images/borboleta.png');
imshow(butterfly);
```

A imagem deve ser modificada para que seja uma borboleta verde com pintinhas rosas. Essa operação foi realizada através de uma máscara que guarda as posições dos pixels com as cores laranja e bege, e substituindo pelas cores pedidas. 

```octave
redChannel = butterfly(:, :, 1);
greenChannel = butterfly(:, :, 2);
blueChannel = butterfly(:, :, 3);

mask1 = redChannel == 255 & greenChannel == 246 & blueChannel == 227;
mask2 = redChannel == 246 & greenChannel == 127 & blueChannel == 0;

redChannel(mask1) = 255;
greenChannel(mask1) = 20;
blueChannel(mask1) = 147;

redChannel(mask2) = 31;
greenChannel(mask2) = 139;
blueChannel(mask2) = 34;

butterflyRGB = cat(3, redChannel, greenChannel, blueChannel);
imshow(butterflyRGB);
```

Por fim a imagem é redimensionada de acordo com o tamanho escolhido pelo usuário.

```octave
width = input("Width: ");
height = input("Height: ");

resizedButterfly = imresize(butterflyRGB, [width, height]);
imwrite(resizedButterfly, './images/resized_butterfly.png');
```
